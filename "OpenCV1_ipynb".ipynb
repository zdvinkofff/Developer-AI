### Импортируем библиотеки для этого задания

```python
import cv2 # импортируем OpenCV
from matplotlib import pyplot as plt # импортируем matplotlib
import numpy as np # импортируем numpy
```

### Упражнение 1
Загрузите цветное изображение [мема](https://drive.google.com/file/d/1c6juvYk3orKzClcQ7rd40eV5WRsIwP4q/view?usp=sharing) и выведите его на экран средствами openCV и matplotlib. Помните, что по умолчанию библиотека openCV цветное изображение воспринимает как BGR (картинка будет выглядеть синей, скорее всего), а не RGB. Для нормального отображения нужно дополнительно преобразовать цветовую схему.

```python
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/sample_data/zhencshiny_37098706_orig_.jpeg', cv2.IMREAD_COLOR)
cv2_imshow(img)
img.shape
```

### Упражнение 2
RGB изображение представляет собой массив, где на каждый пиксель приходится 3 числа. Выведите на экран (можно с помощью matplotlib) раздельно три изображения (значения красного, зеленого и синего каналов).

```python
plt.imshow(img[:, :, 0])
plt.show()
plt.imshow(img[:, :, 1])
plt.show()
plt.imshow(img[:, :, 2])
plt.show()
```

### Упражнение 3
Выведем на экран часть изображения, где содержится лицо персонажа мема. При этом нужно помнить, что можно задавать индексы пикселей: сперва указывается срез по оси У, а потом по оси Х.

```python
img_cut = img[10:270, 150:350]
plt.imshow(img_cut)
plt.show()
```

### Упражнение 4
Функцией resize() увеличьте размер изображения в 2 раза.

```python
img1 = cv2.resize(img_cut, (400, 520))# или размер поменять на (520, 400), тогда изображение будет расстянуто по ширине?
plt.imshow(img1)
plt.show()
```
